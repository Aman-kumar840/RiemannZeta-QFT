```python
# demo.ipynb - Example cells

# 1. Imports
import numpy as np
import matplotlib.pyplot as plt
from classical.riemann_siegel import riemann_siegel
from hybrid.zeta_qft_pipeline import hybrid_zeta

# 2. Classical computation
t_values = [1e2, 1e3, 1e4]
classical_results = [riemann_siegel(t) for t in t_values]

# 3. Hybrid computation
quantum_results = [hybrid_zeta(t, n=8) for t in t_values]

# 4. Plot comparison
plt.plot(t_values, classical_results, label="Classical")
plt.plot(t_values, quantum_results, label="Hybrid Quantum (simulated)")
plt.xlabel("t")
plt.ylabel("Zeta Approximation")
plt.legend()
plt.show()
```
